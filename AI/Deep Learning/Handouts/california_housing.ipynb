# Get Data

```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import fetch_california_housing
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
dataset = fetch_california_housing()
x = dataset.data
y = dataset.target
x_label = dataset.feature_names
y_label = dataset.target_names
```

```python
print(dataset.DESCR)
```

```python
raw_x = pd.DataFrame(x, columns=x_label)
raw_x
```

```python
raw_x.describe()
```

```python
raw_y = pd.DataFrame(y, columns=y_label)
raw_y
```

```python
raw_y.describe()
```

```python
raw_x.plot.box()
```

```python
raw_y.plot.box()
```

# Cleaning Data

```python
x_scaler = MinMaxScaler()
scaled_x = x_scaler.fit_transform(raw_x)
scaled_x = pd.DataFrame(scaled_x, columns=x_label)
```

```python
scaled_x.describe()
```

```python
y_scaler = MinMaxScaler()
scaled_y = y_scaler.fit_transform(raw_y)
scaled_y = pd.DataFrame(scaled_y, columns=y_label)
```

```python
scaled_y.describe()
```

```python
scaled_x.plot.box()
```

```python
scaled_y.plot.box()
```

```python
x_train, x_test, y_train, y_test = train_test_split(scaled_x, scaled_y, test_size=0.3, random_state=5)
```

```python
# x_train
# x_test
# y_train
# y_test
x_train.describe()
# x_test.describe()
# y_train.describe()
# y_test.describe()
```

# Train Model

```python
model = Sequential()
model.add(Dense(200, activation="relu", input_dim=8))
model.add(Dense(1000, activation="relu"))
model.add(Dense(1, activation="linear"))

model.summary()
```

```python
model.compile(optimizer="sgd", loss="mse", metrics=["accuracy"])
```

```python
history = model.fit(x_train, y_train, batch_size=64, epochs=100, verbose=1)
```

```python
fig, loss_ax = plt.subplots()
acc_ax = loss_ax.twinx()

loss_ax.plot(history.history["loss"], "r")
acc_ax.plot(history.history["accuracy"], "g")
plt.show()
```

# Evaluate Model

```python
score = model.evaluate(x_test, y_test, verbose=1)
print("test loss:", score[0])
print("test acc:", score[1])
```

```python
y_pred = model.predict(x_test)
```

```python
plt.plot(y_pred, y_test, "b,")
plt.plot([0, 1], [0, 1], "r--")

plt.grid()
plt.show()
```

```python
inversed_pred = y_scaler.inverse_transform(y_pred)
inversed_actual = y_scaler.inverse_transform(y_test)
```

```python
plt.plot(inversed_pred, inversed_actual, "b,")
plt.plot([0, 5], [0, 5], "r--")

plt.grid()
plt.show()
```
